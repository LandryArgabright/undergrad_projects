```
import pandas as pd
pd.set_option('display.max_rows', 160)
```

```
scrape = pd.read_html('http://en.wikipedia.org/wiki/List_of_triathlon_fatalities')

fatal = scrape[1]

def clean(DataFrame):

  DataFrame.columns = DataFrame.columns.str.lower().str.replace(' ','_')

  DataFrame.date = pd.to_datetime(DataFrame.date)

  DataFrame = DataFrame.dropna()
  DataFrame = DataFrame.drop(columns='name')

  DataFrame['cause/other_notes'] = DataFrame['cause/other_notes'].str.lower()

  DataFrame['event'] = DataFrame.event.str.replace('Swun', 'Swim')
  DataFrame['event'] = DataFrame.event.str.lower().replace('after - next day', 'after')

  return DataFrame
```

```
fatal = clean(fatal)
```

### Viz

```
import plotly.express as px
```

```
fig = px.scatter(fatal,
               x='date',
               y='age',
               hover_data=['country',
                           'cause/other_notes',
                           'triathlon'],
               color='event',
               color_discrete_sequence=px.colors.qualitative.Set1,
               template="plotly_dark",
               title='Age of Death Timeline',
              )
fig.update_layout(
    xaxis_title='Date of Event',
    yaxis_title='Age'
)
fig.show()
```

```
# Separate DataFrames grouped by event column

dt = {'swim':pd.DataFrame,
      'bike': pd.DataFrame,
      'run':pd.DataFrame,
      'later':pd.DataFrame,
      'other':pd.DataFrame,
      'after':pd.DataFrame
      }

for st in fatal['event'].unique():
  dt[st] = fatal[fatal['event'] == st]
tots = pd.concat(dt)
```
