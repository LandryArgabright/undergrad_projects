# Predict Each Stores 90 Days of Sales

```
from fbprophet import Prophet
```

```
# Training on Store 1 Item 1
df = train.query('store == 1 & item == 1')

# Training Set Excludes Nintey days off the end of the said dataframe
train_df = df[:-90]
train_df.columns = ['ds','store','item', 'y']
train_df
```

```
%%time

# Creating an Instance and Fitting Model
m = Prophet(yearly_seasonality=True, weekly_seasonality=True, daily_seasonality=True)
m.fit(train_df[['ds','y']])

# Making Predictions
future = m.make_future_dataframe(periods=90)
demand_forecast = m.predict(future)
```

```
m.plot_components(demand_forecast)
m.plot(demand_forecast)
```

```
# Function Must take 90 days off of the end of each store index
# Must train the model on 90 days less than what each index has
# Must Concat both the prediction dataframes together ( stack together )



# END DATE 
  # 2017-12-31

# TRAINING DATA DATES
  # 2013-01-01 ~ 2017-10-02
# TEST DATA DATES
  # 2017-10-02 ~ 2017-12-31

# START DATE
  # 2013-01-01
```

```
# array of unqiue date
date_array = train['date'].unique()

def pipeline(df):
  
  # instantiating empty dataframes
  prediction_df = pd.DataFrame()
  dataset = pd.DataFrame()

  # selcting each store
  for i in list(range(1,11)):

    # separating stores
    temp_df = df.query("store == {num}".format( num=i )).copy()

    # rename columns - compatible with model
    temp_df['ds'], temp_df['y'] = date_array, temp_df['sales']

    # creating dataframe with all values
    temp_df['ds'] = pd.to_datetime(temp_df['ds'])
    temp_df['year'], temp_df['month'], temp_df['day'] = temp_df['ds'].dt.year, temp_df['ds'].dt.month, temp_df['ds'].dt.day
    temp_df['store'] = i

    # separating train and test data
    train_df, test_df = temp_df[:-90], temp_df[1736:]

    # Creating an Instance and Fitting Model
    m = Prophet(yearly_seasonality=True, weekly_seasonality=True, daily_seasonality=True)
    m.fit(train_df[['ds','y']])
    
    # Making Predictions
    future = m.make_future_dataframe(periods=90)
    demand_forecast = m.predict(future)
    demand_forecast['store'] = i

    # progress output
    print("> " + str(i))
    if i == 10:
      print("> Finished Execution. \n")
      
    # merging all predictions for each store into one dataframe
    prediction_df = prediction_df.append(demand_forecast)

    # merging all data into one dataset ( train and test with additional columns )
    dataset = dataset.append(temp_df)

  return prediction_df, dataset
```

```
forecast_all_stores_df, data_all_stores = pipeline(multi_index_df)
```

```
# drop index of store that caused ambiguity and joined forecasted data and full dataset to allow for comparison
data_all_stores = data_all_stores.droplevel('store')
joined_df = pd.merge(data_all_stores, forecast_all_stores_df, how='inner')
```
